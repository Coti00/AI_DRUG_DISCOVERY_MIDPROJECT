## 1. 질환 선정


### EGFR의 역할과 폐암
EGFR은 세포 성장, 분화, 생존 등을 조절하는 중요한 수용체이다.
폐암에서 EGFR 변이는 치료 타기으로 중요한 역할을 함. 특히 비소세포 폐암(NSCLC)에서 60% 이상이 EGFR을 과발현하고 있기 때문에 EGFR 억제제가 중요한 치료 옵션이 됨

### EGFR 관련 질환의 역학 및 현황
폐암은 전 세계적으로 가장 높은 발병률과 사망률을 보이는 암종 중 하나로, 2022년에는 전 세계적으로 약 250만 건의 새로운 폐암 사례가 발생하고 180만 명이 사망했습니다. 2045년에는 연간 신규 암 발병 건수가 약 3000만 건, 암 관련 사망자 수가 약 1,700만 명으로 예상됩니다[1]

EGFR 돌연변이의 유병률은 인종에 따라 상당한 차이를 보입니다
- 아시아인 폐암 환자의 40-60%에서 EGFR 돌연변이가 발견됩니다.[2]
- 코카시안 인종에서는 10-30% 정도로 상대적으로 낮은 비율을 보입니다[2]

### 현재 치료법과 한계점
#### EGFR 표적 치료제의 발전과 세대별 특성
EGFR 돌연변이 폐암의 치료는 EGFR 타이로신 키나아제 억제제(TKI)를 중심으로 발전해왔습니다. EGFR-TKI는 현재까지 총 3세대가 개발되었습니다:
- 1세대 EGFR-TKI: 게피티닙(Gefitinib), 얼로티닙(Erlotinib) 등이 있으며, PFS 중앙값이 5-7개월 정도로 보고되었습니다.[3]
- 2세대 EGFR-TKI: 아파티닙(Afatinib), 다코미티닙(Dacomitinib) 등이 있으며, PFS 중앙값이 11-14.7개월로 1세대보다 향상된 결과를 보였습니다.[3]
- 3세대 EGFR-TKI: 오시머티닙(Osimertinib), 레이저티닙(Lazertinib) 등이 있으며, 오시머티닙의 경우 PFS 중앙값이 18.9개월로 더욱 향상된 성과를 보였습니다[3]
#### 치료제 내성 발생 메커니즘[4]
##### EGFR 의존적 돌연변이
- T790M돌연변이
   - 메커니즘 : EGFR 유전자 엑손 20의 트레오닌 ➡️ 메티오닌 치환(T790M)으로 ATP 친화력 증가, 약물 결합 방해
   - 임상적 의의 : 1,2세대 EGFR-TKI 내성의 가장 흔한 원인(50 ~ 60%)
- C797S 돌연변이
   - 메커니즘 : EGFR 키나아제 도메인 C797 위치의 시스테인 ➡️ 세린 치환으로 공유결합 형성 불가
   - 임상적 의의 : 오시머티닙 내성의 10-26%에서 관찰
#### 조직학적 변형
- 소세포폐암 전환
  - 메커니즘 : EGFR-돌연변이 선암이 소세포암으로 형질전환되며 EGFR 신호경로와 무관한 증식 메커니즘 획득
  - 임상적 의의 : 내성 환자의 14%에서 발생
#### 비의존적 회피 경로 활성화
- MET 유전자 증폭
  - 메커니즘 : PI3K/AKT 경로 재활성화를 통해 EGFR 신호 대체
  - 임상적 의의 : 오시머티닙 내성의 15-19%에서 발생
- AXL/HER2 과발현
  - 메커니즘 : AXL은 HER2와 이종이합체 형성으로 MAPK/ERK 경로 활성화. HER2 증폭은 EGFR 독립적 신호 전달 유도
  - 임상적 의의 : 내성의 15-20% 기여
- HGF/c-MET경로 활성화
  - 메커니즘 : HGF 발현 증가로 c-MET 수용체 활성화, MEK/ERK경로 회복
  - 임상적 의의 : 교차저항 유발
#### 기타 내성 기전
- IGF-1R 경로 활성화
  - 메커니즘 : IGFBP감소로 PI3K/AKT 경로 재활성화

### 신약 개발의 필요성
1. 높은 유병률과 사망률: 폐암은 전 세계적으로 가장 높은 발병률과 사망률을 보이는 암종이며, 아시아인에서는 EGFR 돌연변이의 비율이 40-60%로 매우 높습니다.[2]
2. 치료제 내성 문제: 현재의 EGFR-TKI는 초기에 높은 반응률을 보이지만, 대부분의 환자가 12개월 내에 내성을 획득하게 됩니다. 이러한 내성 메커니즘을 극복할 수 있는 새로운 치료제가 필요합니다.[6]
3. 개인 맞춤형 치료의 필요성: EGFR 돌연변이의 종류와 내성 메커니즘은 다양하므로, 환자 개인의 유전적 특성에 맞는 맞춤형 치료제 개발이 필요합니다.[2]
4. 부작용 개선의 필요성: 현재 EGFR-TKI는 피부 발진, 손발톱 주위 염증, 심독성 등 다양한 부작용을 유발할 수 있습니다. 효과는 유지하면서 부작용을 줄인 약물 개발이 필요합니다.[3]


EGFR 돌연변이는 특히 아시아인 비소세포폐암 환자에서 높은 비율로 발견되는 중요한 암 발생 원인입니다. 현재 EGFR-TKI를 이용한 표적 치료법이 개발되어 치료 성과가 향상되었지만, 내성 발생이라는 큰 한계에 직면해 있습니다.
이러한 내성 문제를 해결하기 위해 항체약물접합체, 병용요법, 새로운 세대의 EGFR-TKI 등 다양한 접근법이 연구되고 있으며, 최근 아미반타맙과 레이저티닙 병용요법이 기존 치료법보다 우수한 성과를 보여 FDA 우선심사 대상으로 선정되기도 했습니다.
EGFR 돌연변이 폐암 환자의 생존율과 삶의 질을 더욱 향상시키기 위해서는 내성 메커니즘을 극복할 수 있는 새로운 치료제 개발이 지속적으로 필요합니다. 이는 세계적으로 높은 사망률을 보이는 폐암 치료에 있어 중요한 과제이며, 특히 EGFR 돌연변이 비율이 높은 아시아 국가들에서 더욱 시급한 문제입니다.

## 2. 타겟 선정


#### EGFR의 생리적 역할 및 병리적 변이
EGFR(상피세포성장인자수용체)는 세포 성장, 분화, 생존을 조절하는 티로신 키나아제 수용체로, 정상 세포에서는 EGF, TGF-a 등의 리간드 결합 후 이량체화를 통해 신호를 전달함. 이 과정에서 MAPK/ERK 및 PI3K/AKT 경로가 활성화되며, 세포 주기 진행과 혈관 생성 등이 촉진됨.[1]

#### EGFR 변이의 병리학적 메커니즘
1. 리간드 비의존적 활성화
   - 엑손 19 결손 또는 엑손 21 L858R 돌연변이는 EGFR의 키나아제 도메인을 구성적 활성화시켜, 리간드 없이도 지속적인 신호 전달을 유발함. 이로 인해 세포 증식, apoptosis 억제, 전이 촉진이 발생[1]
   - MAPK/ERK 경로 : Ras-Raf-MEK-ERK 연쇄반응을 통해 세포 분열을 촉진
   - PI3K/AKT 경로 : 세포 생존 신호를 강화하며, Bcl-2 계열 단백질을 통해 apoptosis를 억제함[1]
2. 치료 저항성 메커니즘
   - T790M 2차 돌연변이 : 1,2세대 EGFR-TKI 치료 후 50 ~ 60%에서 발생. ATP 친화력 증가로 약물 결합이 방해받아 내성이 유발됨[1]
   - C797S 돌연변이 : 3세대 TKI 오시머티닙의 공유결합 부위 변이로 약물 효능 감소(C707)[2]
   - MET 유전자 증폭 : 15 ~ 19%에서 관찰되며, PI3K/AKT 경로를 대체 활성화 시켜 EGFR 억제제를 회피함[3]
   - 상피-간질 전환(EMT) : 암세포가 이동성과 침습성을 획득하며, AXL 유전자 과발현을 통해 EGFR 신호를 우회함[4]

## 4. 예측 모델 구성


### Molecular fingerprint로 전환

```python
from rdkit.Chem import AllChem
from rdkit.Chem.rdFingerprintGenerator import GetMorganGenerator
import numpy as np

X = []
y = []

# positive dataset
for sm in positive_dataset['Smiles']:
    if isinstance(sm,str):
        m = AllChem.MolFromSmiles(sm)
        gen = GetMorganGenerator(7, fpSize = 1024)
        fp = gen.GetFingerprint(m)
        X.append(np.array(fp))
        y.append(1)

# negative dataset
for sm in negative_dataset['Smiles']:
    if isinstance(sm, str):
        m = AllChem.MolFromSmiles(sm)
        get = GetMorganGenerator(7, fpSize = 1024)
        fp = gen.GetFingerprint(m)
        X.append(np.array(fp))
        y.append(0)

X = np.array(X)
y = np.array(y)
```

### Deep Neural Network를 이용해 예측모델 만들기

```python
import torch
from torch import nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, roc_curve

class DNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(DNN,self).__init__()

        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, 512)
        self.fc3 = nn.Linear(512, output_size)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        self.dropout = nn.Dropout(p=0.6)

    def forward(self,x):
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.sigmoid(self.fc3(x))
        return x
```

```python
model = DNN(input_size=X.shape[1], hidden_size=1024, output_size=1)

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr = 0.01)

X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.1, random_state= 42)

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1,1)
X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.float32).view(-1,1)
```

```python
epochs = 100
train_losses = []
train_accuracies = []
val_accuracies = []

for epoch in range(epochs):
    model.train()

    optimizer.zero_grad()
    outputs = model(X_train_tensor)

    loss = criterion(outputs, y_train_tensor)

    loss.backward()
    optimizer.step()

    with torch.no_grad():
        model.eval()
        predicted = (outputs > 0.5).float()
        correct = (predicted == y_train_tensor).sum().item()
        accuracy = correct / y_train_tensor.size(0)
        train_losses.append(loss.item())
        train_accuracies.append(accuracy)

    with torch.no_grad():
        model.eval()
        val_outputs = model(X_val_tensor)
        val_predicted = (val_outputs > 0.5).float()
        val_correct = (val_predicted == y_val_tensor).sum().item()
        val_accuracy = val_correct / y_val_tensor.size(0)
        val_accuracies.append(val_accuracy)

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss : {loss.item():.4f}, Val Accuracy : {val_accuracy:.4f}')


with torch.no_grad():
    model.eval()
    val_outputs = model(X_val_tensor)
    val_predicted = (val_outputs > 0.5).float()
    val_accuracy = (val_predicted == y_val_tensor).sum().item() / y_val_tensor.size(0)
```

### Roc-Curve

```python
import matplotlib.pyplot as plt

fpr, tpr, thresholds = roc_curve(y_val, val_outputs.detach().numpy())
roc_auc = roc_auc_score(y_val, val_outputs.detach().numpy())

fig = plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

fig.savefig('figures/roc.png')
plt.show()

print(f'Validation Accuracy: {val_accuracy:.4f}')
```
